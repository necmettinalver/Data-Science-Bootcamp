```python
insurance = pd.read_csv('insurance.csv')
```

```python
x_Values = np.linspace(1, 100, 100)
y_Values = np.random.randn(100)

data0 = go.Scatter(
    x = insurance[insurance.smoker=='yes'].bmi,
    y = insurance[insurance.smoker=='yes'].expenses,
    mode = 'markers',
    name = 'Smoker',
    marker = dict(
        size=10,
        color = '#F39C12',
        symbol = 'circle',
        line = dict(width=1)
    )
    )

data1 = go.Scatter(
    x = insurance[insurance.smoker=='no'].bmi,
    y = insurance[insurance.smoker=='no'].expenses,
    mode = 'markers',
    marker = dict(
        size=10,
        color = '#a9dfbf',
        symbol = 'circle',
        line = dict(width=1,color='#0E8700')
    )
    )


layout = go.Layout(
    title = '$Scatter Plot$',
    xaxis = dict(
        title = 'BMI',
        showgrid = False,
        showline = False,
    ),
    yaxis = dict(
        title = '$expenses',
        showgrid = False,
        showline = False,
    ),
)

data=[data0,data1]

fig = go.Figure(data= data,layout=layout)
iplot(fig)
```

```python
from plotly.subplots import make_subplots 

fig = make_subplots(rows = 1, cols = 2, subplot_titles = ('Subplot-1','Subplot-2'))

fig.add_trace(go.Scatter(
    x = x_Values,
    y = y_Values,
    mode = 'markers',
    marker = dict(
        size=10,
        color = 'crimson',
        symbol = 'circle',
        line = dict(width=1,color='#0E8700')
    )),
    row = 1, col = 1
)

fig.add_trace(go.Scatter(
    x = x_Values,
    y = y_Values,
    mode = 'markers',
    marker = dict(
        size=7,
        color = '#DC14CC',
        symbol = 'circle',
        line = dict(width=1,color='#0E8700')
    )),
    row = 1, col = 2
)

fig.update_xaxes(showgrid = False)
fig.update_yaxes(showgrid = False)
fig.show()
```

```python
security = [48, 30, 20, 15]
status = ['Low', 'Medium', 'High', 'Critical']

datax = go.Pie(values = security, labels = status)

layoutx = go.Layout(title = dict(text = 'Güvenlik Düzeyleri',
                                 x = 0.45, y = 0.95, font_size = 20))

fig = go.Figure(data=datax, layout=layoutx)
fig.show()
```

```python
security = [48, 30, 20, 15]
status = ['Low', 'Medium', 'High', 'Critical']
colors = ['#F98DC2','#294827','#BCAFFB',]

datax = go.Pie(values = security, labels = status, marker=dict(colors = colors))

layoutx = go.Layout(title = dict(text = 'Güvenlik Düzeyleri',
                                 x = 0.45, y = 0.95, font_size = 20))

fig = go.Figure(data=datax, layout=layoutx)
fig.show()
```

```python
security = [48, 30, 20, 15]
status = ['Low', 'Medium', 'High', 'Critical']
colors = ['#F98DC2','#294827','#BCAFFB',]

datax = go.Pie(values = security, labels = status, marker=dict(colors = colors),
               textinfo='label+value+percent')#etiketi,değeri,yüzdesi

layoutx = go.Layout(title = dict(text = 'Güvenlik Düzeyleri',
                                 x = 0.45, y = 0.95, font_size = 20))

fig = go.Figure(data=datax, layout=layoutx)
fig.show()
```

```python
security = [48, 30, 20, 15]
status = ['Low', 'Medium', 'High', 'Critical']
colors = ['#F98DC2','#294827','#BCAFFB',]

datax = go.Pie(values = security, labels = status,
               marker=dict(colors = colors,line = dict(color = '#797D7D')),
               textinfo='percent',
               hoverinfo = 'label+value',
               insidetextorientation='radial')#içerdeki metnin açıya göre olmasını sağlıyor

layoutx = go.Layout(title = dict(text = 'Güvenlik Düzeyleri',
                                 x = 0.45, y = 0.95, font_size = 20))

fig = go.Figure(data=datax, layout=layoutx)
fig.show()
```

```python
security = [48, 30, 20, 15]
status = ['Low', 'Medium', 'High', 'Critical']
colors = ['#F98DC2','#294827','#BCAFFB',]

datax = go.Pie(values = security, labels = status,
               marker=dict(colors = colors,line = dict(color = '#797D7D')),
               textinfo='percent',
               hoverinfo = 'label+value',
               insidetextorientation='radial'
               ,pull = [0, 0, 0, 0.2])#hangi parçayı patlatmak istiyorsak onu kullanıyoruz

layoutx = go.Layout(title = dict(text = 'Güvenlik Düzeyleri',
                                 x = 0.45, y = 0.95, font_size = 20))

fig = go.Figure(data=datax, layout=layoutx)
fig.show()
```
