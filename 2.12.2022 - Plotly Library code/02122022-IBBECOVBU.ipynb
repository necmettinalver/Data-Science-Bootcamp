```python
import pandas as pd
import numpy as np
```

```python
#Donut Chart
```

```python
security = [48, 30, 20, 15]
status = ['Low', 'Medium', 'High', 'Critical']
colors = ['#F98DC2','#294827','#BCAFFB',]


data = go.Pie(values = security, labels = status,
               marker=dict(colors = colors,line = dict(color = '#797D7D')),
               textinfo='percent',
               hole=0.4)

layout = go.Layout(title = dict(text = 'Güvenlik Düzeyleri',
                                 x = 0.45, y = 0.95, font_size = 20))

fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
security = [48, 30, 20, 15]
status = ['Low', 'Medium', 'High', 'Critical']
colors = ['#F98DC2','#294827','#BCAFFB',]


data = go.Pie(values = security, labels = status,
               marker=dict(colors = colors,line = dict(color = '#797D7D')),
               textinfo='percent',
               hole=0.4,pull = [0, 0, 0, 0.2])

layout = go.Layout(title = dict(text = 'Güvenlik Düzeyleri',
                                 x = 0.45, y = 0.95, font_size = 20))

fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
suicide = pd.read_csv('suicide.csv')
suicide.head(10)
```

```python
data = go.Pie(
    values = suicide['suicides_no'],
    labels = suicide['age'],
    textinfo='percent',
    hoverinfo = 'label+value'
)

layout = go.Layout(title = dict(text = 'Suicide Data',
                                 x = 0.45, y = 0.95, font_size = 20))


fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
suicide_turkey=suicide[suicide['country']=='Turkey']
```

```python
data = go.Pie(
    values = suicide_turkey['suicides_no'],
    labels = suicide_turkey['age'],
    textinfo='percent',
    hoverinfo = 'label+value'
)

layout = go.Layout(title = dict(text = 'Suicide Data',
                                 x = 0.45, y = 0.95, font_size = 20))


fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
fig = px.pie(suicide, values = 'suicides_no', names = 'generation', title = 'Country')
fig.update_layout(title = dict(text = 'Suicide Data',
                                 x = 0.45, y = 0.95, font_size = 20))
fig.update_traces(textposition = 'outside', textinfo = 'percent+label')
fig.show()
```

```python
#Bubble Chart
```

```python
fig = px.scatter(
    suicide.query('year == 1987'),
    x = 'population',
    y = 'suicides_no',
    size = 'suicides_no',
    color = 'age',
    hover_name = 'country',
    size_max = 60
)

fig.update_xaxes(showgrid = False)
fig.update_yaxes(showgrid = False)
fig.show()
```

```python
insurance = pd.read_csv('insurance.csv')
insurance.head()
```

```python
fig = go.Figure(data=[go.Table(
    header = dict(
        values = list(insurance.columns),
        line_color = 'black',
        fill_color = 'orange',
        align = 'center',
        height = 40,
        font = dict(color = 'white', size = 18)
    ),
    cells = dict(values = [
        insurance.age,
        insurance.sex,
        insurance.bmi,
        insurance.children,
        insurance.smoker,
        insurance.region,
        insurance.expenses
    ],
            line_color = 'darkgrey',
            fill_color = 'lightcyan',
            align = 'left')
)])
fig.show()
```

```python
fig = ff.create_table(insurance.tail(10))
fig.show()
```

```python
colorscale = [[0,'red'],[.5, '#DCE775'],[1,'#FFDCAF']]
fig = ff.create_table(insurance.tail(10),colorscale=colorscale)
fig.show()
```

```python
fig = px.sunburst(insurance, path = ['region','smoker','sex'], values='expenses')
fig.update_layout(height=800,width=800)
fig.show()
```
