```python
#25.11.2022 tarihinden sonraki kısımlar
x=np.array([1,2,3,4,5,6])
w=np.tan(x)
e=np.tanh(x)
y=np.power(x,3)
z=np.power(x,2)
d=np.sin(x)
c=np.sin(x)
f=np.cos(x)
z=np.cosh(x)

fig1,ax=plt.subplots(nrows=3,ncols=3,figsize=(12,8))

ax[0,0].plot(x,w,linestyle='dashed')
ax[0,1].plot(x,e)
ax[0,2].plot(x,d)

ax[1,0].plot(x,c)
ax[1,1].plot(x,f)
ax[1,2].plot(x,z)

ax[2,0].plot(x,y)
ax[2,1].plot(x,w)
ax[2,2].plot(x,z)

plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.power(x,2)
y2 = np.power(x,3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)

fig1, ax1 = plt.subplots(nrows=3, ncols=3, figsize=(14,14))

calcs = (y1,y2,y3,y4,y5,y6,y7,y8,y9)
titles = ('x^2','x^3','sin(x)','cos(x)','tan(x)','tanh(x)','sinh(x)','cosh(x)','exp(x)')
colors = ("red","blue","green","magenta","lightblue","purple","lightgreen","orange","black")

for temp, ij in enumerate([(i,j) for i in range(3) for j in range(3)]):
    i,j=ij[0],ij[1]
    
    # calculations and labels (for legend)
    ax1[i,j].plot(x,calcs[temp], c=colors[temp], linewidth=3, label=titles[temp])
    
    # titles
    ax1[i,j].set_title(titles[temp], fontsize=14)

    # x and y labels
    ax1[i,j].set_xlabel("x")
    ax1[i,j].set_ylabel("y")
    
    # adding grid
    ax1[i,j].grid()
    
    # to carry labels to perimeter
# for ax in ax1.flat:
#     ax.label_outer()
    
fig1.suptitle("SOME CALCULATIONS", fontsize=25)
fig1.supxlabel("X values", fontsize=18)
fig1.supylabel("Y values", fontsize=18)
fig1.legend(loc='upper left', fontsize=14, bbox_to_anchor=(-0.07, 0.93))
fig1.tight_layout(pad=2)

plt.show()
```
